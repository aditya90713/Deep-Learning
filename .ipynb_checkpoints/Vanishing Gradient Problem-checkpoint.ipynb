```python
import pandas, numpy
import matplotlib
%matplotlib inline
import tensorflow
import keras
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from keras.layers import Dense
from keras.models import Sequential
```

```python
#!pip install tensorflow
```

```python
x, y=make_moons(n_samples=250, noise=0.05, random_state=42)
```

```python
x.shape
```

```python
y.shape
```

```python
matplotlib.pyplot.scatter(x[:,0],x[:,1], c=y, s=100)
matplotlib.pyplot.show()
# matplotlib.pyplot.legend()
```

```python
model = Sequential()

# using Relu activation fucntion in hidden layers
#input layer
model.add(Dense(10, activation='sigmoid', input_dim=2))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
# output layer or node
model.add(Dense(1, activation='sigmoid'))




#input layer
#model.add(Dense(10, activation='sigmoid', input_dim=2))
#model.add(Dense(10, activation='sigmoid'))
#model.add(Dense(10, activation='sigmoid'))
#model.add(Dense(10, activation='sigmoid'))
# 1. reducing model complexity using less number of hidden layer
#model.add(Dense(10, activation='sigmoid'))
#model.add(Dense(10, activation='sigmoid'))
#model.add(Dense(10, activation='sigmoid'))
#model.add(Dense(10, activation='sigmoid'))
#model.add(Dense(10, activation='sigmoid'))
#model.add(Dense(10, activation='sigmoid'))
#model.add(Dense(10, activation='sigmoid'))
# output layer or node
#model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# we can randomly initialized weights
model.get_weights()
```

```python
# first layer weights
old_Weights = model.get_weights()[0]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
x_train
```

```python
# we train our model for one epoch or one iteration
#model.fit(x_train, y_train, epochs=1)

# now we can train model more number of epoch epoch=100
model.fit(x_train, y_train, epochs=1)
```

```python
# here we can see change in weights
new_weights = model.get_weights()[0]
```

```python
model.optimizer.get_config()["learning_rate"]
```

```python
# here we calculate gradient means change in weights 
gradient = (old_Weights-new_weights)/0.001
```

```python
percent_change= abs(100*(old_Weights-new_weights)/old_Weights)
```
